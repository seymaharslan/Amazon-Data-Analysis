# Interpretation of Sales Analysis

As the outcome of Sales analysis on R, it showed us there is a Linear Relationship between Views and Quantity Sold.
This means more views more sales. So, how do we get more views?
To be able to answer this question, we need to run another analysis with the related variables. 

### What does customer see?

When searching a keyword on Amazon website, below is an example of what we see. Red marked areas will be our variables to analyze. 

## Meta Data


* Price: Item price
* Views: Total number of clicks recieved 
* Ratings: Average product evaluation score given by customer for the the particular item 
* Content Score: Product page content quality score out of 10
* Review Count: Number of feedbacks received for the particular item

## Goal

We want to see which variable has the significant effect to a customer for them to click on the product page

## Method


* Stepwise Multiple Linear Regression
* IV: Price, Review Count, Ratings, Content
* DV: Views
